### 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker')
```

### 2. BarPlot (Numerical - Categorical)

```python
sns.barplot(y='tip', x='sex', data=tips, hue='smoker')
```

```python
titanic.head()
```

```python
sns.barplot(y='Age', x='Pclass', hue='Sex', data=titanic)
```

```python
(titanic['Pclass'] == 3).value_counts()
```

### 3. BoxPlot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['Sex'],y= titanic['Age'], hue=titanic['Survived'])
```

### 4. DistPlot (Numerical - Categorical)

```python
sns.displot(titanic[titanic['Survived'] == 0]['Age'], kind='kde')
sns.displot(titanic[titanic['Survived'] == 1]['Age'], kind='kde', color='red')
```

### 5. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
titanic.head(2)
```

### 5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```
