```python
from idlelib.redirector import OriginalCommand

from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car', as_frame=True).frame
data
```

```python
# Encode the columns into integer from string.
# small = 0, med = 1, high = 2
columns_to_encode = ['lug_boot', 'safety']

encoder = OrdinalEncoder(
    categories= [
        ['small', 'med', 'big'],
        ['low', 'med', 'high'],
    ]
)
data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
data
```

```python
encoder.inverse_transform(data[columns_to_encode])   # inverse the transforms
```

```python
# inverse the transforms into the same columns
data[columns_to_encode]=encoder.inverse_transform (data[columns_to_encode])
data
```

```python
data = fetch_openml('adult',as_frame=True).frame
data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data.occupation.value_counts()
```

```python
import pandas as pd
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoded_values = encoder.fit_transform(data[['occupation', 'race']])
new_cols = encoder.get_feature_names_out(['occupation', 'race'])
```

```python
encoded_values
```

```python
new_cols
```

```python
df_encoded = pd.DataFrame(encoded_values, columns=new_cols, index = data.index)

data_final = pd.concat([data.drop(columns=['occupation', 'race']),df_encoded], axis=1)
```
